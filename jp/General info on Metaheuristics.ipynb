### Optimizacion
Se basa en mejorar tanto como sea posible una funcin objetivo, esto es:
Maximizar o minimizar una variable, segn el problema.
El problema suele radicar en *maximizar* ganancias o en *minimizar* costos basados en un set de restricciones, que suelen ser recursos disponibles para resolver dicho problema.

### Espacio de bsqueda
Se refiere al conjunto de todas las soluciones posibles para un problema.

Este puede ser infinito, en el caso de problemas contnuos, o finito en caso de problemas combinatorios.
#### Explotacin
Consiste en enfocarse en buscar mejores soluciones en un rango acotado del espacio de bsqueda.

Este enfoque suele dar como resultado el encontrar satisfactoriamente mnimos y mximos, pero existe la posibilidad de que sean "locales", y por tanto no la solucin ptima al problema.
#### Exploracin
Consiste en enfocarse en explorar globalmente el espacio de bsqueda, lo que permite encontrar una cantidad mas diversa de soluciones, pero no necesariamente encontrando estos mnimos locales para cada una de las soluciones planteadas.

### Tipos de bsqueda
#### Busqueda local
Este tipo de busqueda se enfoca en recorrer el vecindario dando 1 paso a la vez.

Es muy efectivo para encontrar mnimos/mximos locales, pero suele tener el efecto negativo de ciclarse en dichos mnimos/mximos.

Es por esto, que las metaheursticas de bsqueda local, involucran conceptos que los hacen salir de su "zona de confort", ya sea forzando el salir del vecindario mediante perturbaciones (Al hablar de mtodos combinatorios), o bien permitiendo la toma de "malas decisiones" para fomentar la exploracin del espacio de solucin.

##### Paso
Criterio de decisin para elegir una solucin a partir de una previa, por definicin, acota el rango de posibles soluciones siguientes al "vecindario" de la solucin.

##### Vecindario
Es el subconjunto de soluciones al que se puede llegar en un solo paso.
* En problemas combinatorios, tiene un tamao definido y finito.
* En problemas continuos, puede tener a su vez un tamao infinito, acotado mas bien por la resolucin de la solucin y por intervalos dados a las variables de entrada.


#### Busqueda global
Su enfoque radica en explorar todo el espacio de bsqueda, por lo cual no aplica un concepto de vecindario o paso, sino que simplemente genera nuevas soluciones vlidas dentro del espacio de bsqueda. Estas soluciones pueden estar correlacionadas con la solucin anterior, pero no necesariamente en un concepto de "cercana" (Vecindario), sino de "semejanza".

### Tipos de problemas
#### Continuos
Son problemas cuyo objetivo es minimizar o maximizar el valor objetivo de una funcin continua. Dicha funcin puede ser n-dimensional (f(x), f(f, y), f(x, y, z...)).

Este tipo de funciones, se trabajan discretizandose, haciendose muy relevante el concepto de "Vecindario" y "Paso", pues definirn tanto la resolucin de la solucin, como el tiempo que tomar llegar a ella.



#### Combinatorios
Son problemas cuya solucin es de la forma [A, B, C, D...] (Una combinacin), tienen la caracterstica de tener una cantidad finita de soluciones, pero suelen ser tantas que no sera prctico iterarlas todas a fin de encontrar la mejor solucin.




## Simulated annealing
#### Introduccion
Se trata de un mtodo que modela de forma simple el proceso de recocido de metales (Calentamiento de un material y posterior enfriamiento para reducir defectos, efectivamente minimizando la energa del sistema).

#### Caractersticas
Neighborhood-based search.

Recorre vecindario un paso a la vez, tendiendo naturalmente a dirigirse a las zonas mnimas locales en una forma similar al mtodo de gradiente descendiente.

Permite la salida de mnimos locales aprovechando el concepto de 'Energa', el cul es una medida de la disposicin del sistema a moverse a zonas menos ptimas a fin de fomentar la exploracin de la solucin.

Util tanto en problemas continuos como en combinatorios.

#### En problemas continuos
Se define un paso como la seleccin de una solucin de forma aleatoria, acotada en un intervalo alrededor de la solucin actual.

#### En problemas combinatorios
Se define un paso como el intercambio de 2 elementos del conjunto o secuencia Solucin, dando lugar a una nueva solucin.


## Genetic algorithms

#### Introduccion
Pertenece a la rama de los <b>Algoritmos evolutivos</b>.
Inspirado por conceptos de la naturaleza y la evolucin:
* Seleccin natural
* Reproduccin
* Supervivencia del mas apto

ms conceptos adoptados:

 * Cruces genticos

On reproduction:
Puede realizarse mediante diferentes mecanismos
* Roulette wheel
 * Se eligen los padres como si se girara una ruleta, teniendo mayor probabilidad los padres que tengan mejor valor objetivo.
* Tournament selection

#### Caractersticas
Population-based search (Global search)

Binary encoding (Trabaja con un resultado de la forma **[0,1,1,0,0,1,1...]** conocido como cromosoma).

El encoding del resultado se conoce como **Genotipo**

El decoding del resultado (Su significado) se conoce como **Fenotipo**

Cada elemento de la solucin, se conoce como **Alelo**, el cual contiene un **gen** (0 o 1)

Mutaciones: Provocan alteraciones aleatorias en el cromosoma que representa la solucin (Aunque la probabilidad de que stas ocurran suele ser muy baja).

#### En problemas continuos

Se requiere definir la resolucin de la solucin a fin de poder acotarla a un cromosoma de longitud finita.

Para obtener el valor de la solucin, se debe pasar por un proceso de decoding/encoding de los cromosomas.

#### En problemas combinatorios

No existe el proceso de encoding/Decoding, pues el cromosoma es directamente el conjunto representando la solucin.

Cambian en gran medida las implementaciones de los conceptos como Mutacin y reproduccin, pues stos fenomenos deben evitar el repetir elementos en la solucin final.

## Tabu search
#### Introduccion
Metaheurstica que emplea mtodos de busqueda local para sus optimizaciones.
Se basa en el concepto de **Tab**, palabra Tongan que significa "Lo que no debe ser tocado, debido a ser sagrado".

Es el padre de una familia de approaches derivados que introducen estructuras de memoria a las metaheursticas.

#### Caracteristicas
Lista Tab: Lleva registro de las ultimas soluciones exploradas, para evitar caer en ellas y ciclarse en mnimos locales o soluciones cclicas. Esto es efectivamente una memoria de corto plazo.

Criterio de aspiracin: Criterio que hace **Override** al status de tab de cierta solucin, sto para evitar ciclos o convergencia prematura.


#### En problemas continuos
No aplica para problemas continuos, pues el concepto de memoria est limitado a una sola solucin, por lo que carecera de utilidad en un problema con soluciones infinitas.

Otro problema radica en tener que conocer el vecindario completo y ordenarlo en base a su valor objetivo, pues a su vez dara pie a un vecindario de tamao infinito.
